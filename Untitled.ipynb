```python
import os
import json
from sage.all import GF, vector, matrix

# --- Configuración de Parámetros ---
input_dir = "generated_data"
filename = "instances_n7_k3_r2_m4.json"
input_path = os.path.join(input_dir, filename)

# Índice de la instancia de prueba (0 a 19 para cubrir las 20 instancias)
idx = 19 

# --- Cargar archivo JSON ---
if not os.path.exists(input_path):
    raise FileNotFoundError(f"No se encontró el archivo: {input_path}")

with open(input_path, "r") as f:
    dataset = json.load(f)

# --- Recuperar Parámetros Globales ---
params = dataset["parameters"]
q_val, n, k, r, m = params["q"], params["n"], params["k"], params["r"], params["m"]
n_minus_k = n - k

print(f"--- Cargando Configuración: n={n}, k={k}, r={r}, m={m} ---")

# --- Instanciar Entorno Algebraico en Sage ---
Fq = GF(q_val)
Fqm = GF(q_val**m, name='z')
V_Fq = Fqm.vector_space(map=False)  # Espacio vectorial F_q^m

# --- Extraer la Instancia en el Índice Seleccionado ---
instance = dataset["instances"][idx]
instance_id = instance["instance_id"]
print(f"Verificando Instancia ID: {instance_id} (Índice en JSON: {idx})")

# --- Reconstrucción Algebraica (Mapeo de Bits a Cuerpos Finitos) ---

# 1. Reconstruir la matriz H (Tamaño: (n-k) x n)
H_rows = []
for i in range(n_minus_k):
    row_elements = [Fqm(V_Fq(vector(Fq, instance["H"][i][j]))) for j in range(n)]
    H_rows.append(row_elements)
H = matrix(Fqm, H_rows)

# 2. Reconstruir el vector Syndrome original s (Tamaño: (n-k))
s = vector(Fqm, [Fqm(V_Fq(vector(Fq, instance["s"][i]))) for i in range(n_minus_k)])

# 3. Reconstruir el vector de error secreto e (Tamaño: n)
e_secret = vector(Fqm, [Fqm(V_Fq(vector(Fq, instance["e"][j]))) for j in range(n)])

# --- Operación y Validación Algebraica ---
s_calculated = H * e_secret
match = (s_calculated == s)

# --- Impresión de Resultados ---
print("\n--- Resultados de la Validación ---")
print(f"Vector de error secreto e: {e_secret}")
print(f"Síndrome original s:      {s}")
print(f"Síndrome calculado H*e:    {s_calculated}")
print(f"\n¿Coincide H * e == s?: {match}")

if match:
    print("[✓] ÉXITO: La reconstrucción algebraica es correcta y cumple con la ecuación de paridad.")
else:
    print("[X] ERROR: El síndrome calculado no coincide con el original.")
```
